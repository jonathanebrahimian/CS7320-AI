## PEAS description of the cleaning phase

__Performance Measure:__ Each action costs 1 energy unit. The performance is measured as the sum of the energy units used to clean the whole room.

__Environment:__ A room with $n \times n$ squares where $n = 5$. Dirt is randomly placed on each square with probability $p = 0.2$. For simplicity, you can assume that the agent knows the size and the layout of the room (i.e., it knows $n$). To starte, the agent is placed on a random square.

__Actuators:__ The agent can `clean` the current square or move to an adjacent square by going `north`, `east`, `south`, or `west`.

__Sensors:__ Four bumper sensors, one for north, east, south, and west; a dirt sensor reporting dirt in the current square.  


## The agent program for a simple randomized agent

The agent program is a function that gets sensor information (the current percepts) as the arguments. The arguments are:

* A dictionary with boolean entries for the for bumper sensors `north`, `east`, `west`, `south`. E.g., if the agent is on the north-west corner, `bumpers` will be `{"north" : True, "east" : False, "south" : False, "west" : True}`.
* The dirt sensor produces a boolean.

The agent returns the chosen action as a string.

Here is an example implementation for the agent program of a simple randomized agent:  

```python
from numpy import random

actions = ["north", "east", "west", "south", "suck"]


def simple_randomized_agent(bumpers, dirty):
    return random.choice(actions)
```

```python
simple_randomized_agent({"north" : True, "east" : False, "south" : False, "west" : True}, True)
```

__Note:__ This is not a rational intelligent agent. It ignores its sensors and may bump into a wall or not clean a dirty square. You will be asked to implement rational agents below.

## Simple environment example

This simple environment is infinite in size (bumpers are always `False`) and every square is always dirty, even if the agent cleans it. The environment function returns the performance measure which is here the number of cleaned squares (since all squares are constantly dirty, it is the number of `suck` actions by the agent). 

```python
def simple_environment(agent, max_steps, verbose = True):
    num_cleaned = 0
    
    for i in range(max_steps):
        dirty = True
        bumpers = {"north" : False, "south" : False, "west" : False, "east" : False}

        action = agent(bumpers, dirty)
        if (verbose): print("step", i , "- action:", action) 
        
        if (action == "suck"): 
            num_cleaned = num_cleaned + 1
        
    return num_cleaned
        
```

Do one simulation run with 20 steps.

```python
simple_environment(simple_randomized_agent, max_steps = 20)
```

## Task 1: Implement a simulation environment [2 Points]

The simple environment above is not very realistic. Your environment simulator needs to follow the PEAS description from above. It needs to:

* Initialize the environment by storing the state of each square (clean/dirty) and making some dirty.
* Keep track of the agent's position.
* Call the agent function repeatedly and provide the agent function with the sensor inputs.  
* React to the agent's actions. E.g, by removing dirt from a square or moving the agent around unless there is a wall in the way.
* Keep track of the performance measure. That is, track the agent's actions until all dirty squares are clean and count the number of actions it takes the agent to complete the task.

The easiest implementation for the environment is to hold an 2-dimensional array to represent if squares are clean or dirty and to call the agent function in a loop until all squares are clean or a predefined number of steps have been reached (i.e., the robot runs out of energy).

The simulation environment needs to work with the simple randomized agent program from above and then it can be used for your agent implementation in the tasks below.

```python
#imports
import random
import pandas as pd
import numpy as np
import copy
```

```python
#print a room with linebreaks
def pretty_print(floor):
    for x in floor:
        print(x)

#return True if no squares are dirty
def done(floor):
    for x in floor:
        if 1 in x:
            return False
    return True
```

```python
# Your code and description goes here
save_room = [] #global storage of room
def simmulate(agent,size=5,max_steps = 30, new_room = True, verbose = False):
    global save_room

    
    #initialize new room or load old room
    if new_room:
        room = [[1 if random.random() < .2 else 0 for i in range(size)] for i in range(size)]
        save_room = copy.deepcopy(room)
    else:
        room = copy.deepcopy(save_room)
    
    tracker = [[0 for i in range(size)] for i in range(size)]
    

    points = 0 #store points
    
    x = random.randint(0, size-1) #random initial x position
    y = random.randint(0, size-1) #random initial y position
    tracker[y][x] = 1
    
   
        
    
    for itteration in range(max_steps):

        bumpers = {"north" : y == 0, "east" : x == size-1, "south" : y == size-1, "west" : x == 0} #setting bumper sensors
        move = agent(bumpers, room[y][x]) # call agent function
    
        if verbose:
            print("before move")
            print("x: " + str(x))
            print("y: " + str(y))
            print("move: " + move)
            pretty_print(tracker)
  
        #perform move
        if move == "suck":
            room[y][x] = 0
            points += 1
        elif move == "west" and not bumpers["west"]:
            x += -1
        elif move == "east" and not bumpers["east"]:
            x += 1
        elif move == "north" and not bumpers["north"]:
            y += -1
        elif move == "south" and not bumpers["south"]:
            y += 1
            
            
        tracker[y][x] = 1
        points += 1 #calculate points

        #return number of points when the whole room is clened
        if done(room):
            return points
    
    
    return -1

```

For my environment I create a room using a 2D array of a given size and assign 0 to a clean square and 1 to a dirty square. As the description requests, there is a 20% chance that a square is dirty when the room is initialized. I have added a global variable called save_room that contains a deep copy of a newly generated room in case I want to run multiple agents on the same room. After initializing the room, a random start point is decided and the simulation begins. The bumper sensor values are calculated and sent to the agent function along with the a 0 or 1 denoting if the current square is dirty. The agent function then returns a move to make and the environment performs the move (as long as the move does not make the robot run into a wall). The environment then stops if the room is clean or continues if it is dirty. Once the room is cleaned the environment will return the number of actions the robot performed. If the room is not cleaned by the end of the max_steps, a -1 is returned.

## Task 2:  Implement a simple reflex agent [1 Point] 

The simple reflex agent randomly walks around but reacts to the bumper sensor by not bumping into the wall and to dirt with sucking.

```python
import random
def reflex_agent(bumpers, dirt):
    #suck if dirt is detected
    if dirt:
        return "suck"
    actions = [key for key,value in bumpers.items() if not value] #get all available moves
    return random.choice(actions) #return random choice
```

The simple reflex agent performs a random walk while not bumping into walls and sucks if dirt is detected.

## Task 3: Implement a model-based reflex agent [3 Point]

This agent keeps track of the location and remembers where it has cleaned. Assume the agent knows how many squares the room has. It can move to a corner to determin its location and then is able to use more advanced navigation.

_Note on implementing the state:_ You can use a global variable. In Python, you have to use the keyword `global` in your function for this to work (see: https://www.programiz.com/python-programming/global-keyword). Alternatively, you can define a class for your agent with a member variable for the state and a function for the agent program (see: https://www.w3schools.com/python/python_classes.asp). 

Describe how you define the __agent state__ and how your agent works before implementing it.

I initially wanted the robot to only go over each square only once because this would be the optimal solution. Unfortunately, after numerous attempts, I decided that there are too many edge cases due to random start positions and that it would not be possible without having GPS or vision percepts. I then decided too create a two step process that is not optimal but an efficient compromise. Step one of the cleaning process is moving the robot to the top left of the room. This is done by moving north until the northern bumper sensor is triggered then moving west until the west bumper sensor is triggered. Step two of the cleaning process is snaking from top to bottom. This ensures that once the robot hits the top left corner, we take an optimal path to vacuum each square once. This approach is also memory efficient, the robot only needs to store a boolean to remember if the robot has finished step one and a string to remember the last move that was made.

The worst case scenario for this robot is if it randomly starts in the bottom right corner. This is because it will need to drive all the way across the room to get to the starting position in the top left corner maximizing the number of squares that are driven over twice.

The best case scenario for this robot is if it randomly starts in the top left corner. This would make our robot perform the optimal solution with the percepts given because we would never drive over the same square twice.

```python
# Your code goes here
# Your code and description goes here
class Smart_Agent:
    def __init__(self, size):
        self.corner = False
        self.size = size
        self.last_move = ""
    

    def move(self,bumpers,dirty):
        #if dirty suck
        if dirty:
            return "suck"
        
        actions = [key for key,value in bumpers.items() if not value] #get possible actions
        
        #Get to top left corner
        if not self.corner and "north" in actions:
            self.last_move = "north"
            return "north"
        elif not self.corner and "west" in actions:
            self.last_move = "west"
            return "west"
        elif not self.corner:
            self.last_move = "east"
            self.corner = True
            
        
        #perform snaking maneuver 
        if "east" in actions and (self.last_move == "east" or self.last_move =="south"):
            self.last_move = "east"
            return "east"
        elif "west" in actions and (self.last_move == "south" or self.last_move == "west"):
            self.last_move = "west"
            return "west"
        elif "south" in actions and (self.last_move == "east" or self.last_move == "west"):
            self.last_move = "south"
            return "south"
        
        return "stuck" #if reached, the robot would be stuck and never return a move
        
```

## Task 4: Simulation study [3 Points]

Compare the performance of the agents using different size environments. E.g., $5 \times 5$, $10 \times 10$ and
$100 \times 100$. Use at least 100 random runs for each. Present the results in a suitable format (tables, graphs) and discuss the differences.

```python
# Your code goes here
random_agent_scores = []

simple_randomized_scores = []
simple_reflex_scores = []
mode_based_scores = []
size_save = []

for size in [5,10,100]:
    for run in range(100):
        simple_randomized_scores.append(simmulate(simple_randomized_agent, size,max_steps= size ** size * 4))
        simple_reflex_scores.append(simmulate(reflex_agent, size,max_steps= size ** size * 4,new_room=False))
        mode_based_scores.append(simmulate(Smart_Agent(size).move, size,max_steps= size ** size * 4,new_room=False))
        size_save.append(size)
        if size == 0 and run % 10 == 0:
            print('Finished ', run, ' of size 100')
    
    print("Finished room of size: ",size)
    

performance = {
    'randomized':simple_randomized_scores,
    'simple_reflex':simple_reflex_scores,
    'model_based':mode_based_scores,
    'size':size_save
}

```

All agents were put through the same 100 randomized rooms for each size room. The max steps was decided randomly by the equation size ^ size * 4.

```python
# build a room with obstacles
# create a padding layer of walls for the room for reachable detection
# 5% chance of obsticle persquare
# 20% chance of dirt per square
def build_object_room(size):
    valid_room = False
    while not valid_room:
        room = [[-1 for x in range(size+2)]]
        for i in range(size + 1):
            temp = [-1]
            for x in range(size):
                rng = random.random()
                if rng < .05:
                    temp.append(-1)
                elif rng < .25:
                    temp.append(1)
                else:
                    temp.append(0)
            temp.append(-1)
            room.append(copy.deepcopy(temp))

        room.append([-1 for x in range(size+2)])
        
        #ensure room does not have unreachable dirt
        if reachable(room,size):
            valid_room = True
    

    return room
        

#Build a room tracker for debug purposes
def track_room(room):
    for row in room:
        y = 0
        for x in row:
            if x == 1:
                row[y] = 0
            y += 1
    return room

#make sure objects do not enclose a dirty square
def reachable(room,size):
    for x in range(0,size):
        for y in range(0,size):
            if room[y+1][x+1] == 1 and room[y+2][x+1] == -1 and room[y+1][x+2] == -1 and room[y][x+1] == -1 and room[y+1][x] == -1:
                return False
    return True
                      
```

Below are helper functions for creating the room with objects. There is a 5% chance per square that an object is placed and a 20% chance per square that dirt is placed. One big adjustment that was made to the room is that I added 2 rows and 2 columns to the room to create outer walls. This was done in order to be able to check if all dirt spots are reachable. For example: if there is a dirt spot in the top right corner of the room, I want to make sure that it is not enclosed by objects therefor making it unaccessible. This check was done by going through every square in the room and ensuring that one of the north, east, south, or west squares is not a wall or an object (denoted by a -1).

```python
def simmulate_object(agent,size=5,max_steps = 30,new_room=True, verbose = False):
    global save_room
    
    #initialize new room or load old room
    if new_room:
        room = build_object_room(size)
        save_room = copy.deepcopy(room)
    else:
        room = copy.deepcopy(save_room)
    
    if verbose:
        pretty_print(room)
    tracker = track_room(copy.deepcopy(room))
    

    points = 0 #store points
    good_start = False
    
    #make sure we do not start on an object
    while not good_start:
        x = random.randint(0, size-1) #random initial x position
        y = random.randint(0, size-1) #random initial y position
        
        if room[y+1][x+1] != -1:
            tracker[y+1][x+1] = 1
            good_start = True
    
    for itteration in range(max_steps):

        bumpers = {"north" : room[y][x+1] == -1, "east" : room[y+1][x+2] == -1, "south" : room[y+2][x+1] == -1, "west" : room[y+1][x] == -1} #setting bumper sensors
        move = agent(bumpers, room[y+1][x+1]) # call agent function
    
        if verbose:
            print("before move")
            print("x: " + str(x))
            print("y: " + str(y))
            print(bumpers)
            print("move: " + move)
            pretty_print(tracker)
  
        #perform move
        if move == "suck":
            room[y+1][x+1] = 0
            points += 1
        elif move == "west" and not bumpers["west"]:
            x += -1
        elif move == "east" and not bumpers["east"]:
            x += 1
        elif move == "north" and not bumpers["north"]:
            y += -1
        elif move == "south" and not bumpers["south"]:
            y += 1
            
            
        tracker[y+1][x+1] = 1
        points += 1 #calculate points

        #return number of points when the whole room is clened
        if done(room):
            return points
        
        if move == "stuck":
            if verbose:
                print("Could not clean full room. Robot is stuck.")
            return -1
    
    
    return -1
```

Also, the detection for bumper sensors was changed due to the new representation of the room. To minimize execution time, I added functionality for the robot to detect and tell the environment if it is stuck. If the robot is stuck, the run is deemed unsuccessful and -1 is returned.

### Changes

The changes I would make to the model based agent would primarily be in the snaking down portion of the algorithm (after the robot has already reached the top left corner). When the agent is moving east or west and runs into an obstacle I would go south then check if I can continue to go the same horizontal direction I was before. If yes I would continue to go that direction and after each step I would try to go north if possible and continue in the horizontal direction. This would allow me to maneuver around singular objects in the middle of the room. For example, if I was moving east and I encounter a singular object to the east, I would go: south, east, north, east, etc. and avoid the object. This would require me to use more memory to remember not only the previous move I made but also the main direction I am navigating in. 

Another change I would make would be that if I can no longer go south during the snaking phase, I would not declare my robot to be stuck. Instead, I would do the opposite of my previous move until I could go south. Therefore, if my robot encounters the eastern wall but there is an object preventing it from moving south, I would perform one move to the west (opposite of east) then go south then continue to go the opposite direction from the wall (west). This would again require more storage to signify that this backup maneuver is starting and to store the multiple previous moves.

These changes would allow my model based agent to have more of a chance to clean full rooms.
